# 4. Graphical Models

```python
import numpy as np
import matplotlib.pyplot as plt

import torch
import torch.nn.functional as F
```

```python
# 4.2.1 Representing the joint distribution

# unstructured joint probs require O(K^NG) params to describe the join prob configuration
# whereas structured joint probs only require O(NG*K^(Np+1))
K = np.linspace(1, 30, 30)
Ngf = 50
Ng = np.linspace(1, Ngf, Ngf)
Np = np.linspace(1, Ngf, Ngf)

p_uns = lambda ki, ng: ki**ng
p_str = lambda ki, ng, np: ng*ki**(np+1)

# varying K
fig, ax = plt.subplots(1, 2, figsize=(15,6))
ax[0].plot(K, p_uns(K, Ng[-1]), label="O(K^NG)")
ax[0].plot(K, p_str(K, Ng[-1], Np[1]), label="O(NG*K^(Np+1))")
ax[0].set_xlabel("K")
ax[0].set_ylabel("Order")
ax[0].set_title("varying K")
ax[0].legend()
fig.tight_layout()

# varying NG
ax[1].plot(Ng, p_uns(K[-1], Ng), label="O(K^NG)")
ax[1].plot(Ng, p_str(K[-1], Ng, Np[1]), label="O(NG*K^(Np+1))")
ax[1].set_xlabel("NG")
ax[1].set_ylabel("Order")
ax[1].set_title("varying NG")
ax[1].legend()
fig.tight_layout()
```
